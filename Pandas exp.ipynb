```python
import numpy as np
uber_data['MILES_3'] = np.where(uber_data['MILES*'] <= 100, "short trip", 
                                np.where((uber_data['MILES*'] > 100) & (uber_data['MILES*'] < 200), "medium trip", "long trip"))
uber_data
```

```python
conditions=[(uber_data['MILES*']>200),(uber_data['MILES*']>100)&(uber_data['MILES*']<=200),(uber_data['MILES*']<=100)]
categories=["Long trip","Medium trip","Short trip"]
uber_data['MILES_CAT1']=np.select(conditions,categories,default="Unknown")
uber_data
```

```python
uber_data=uber_data.drop(['MILES_3'],axis=1)
uber_data
```
